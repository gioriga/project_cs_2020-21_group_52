### 1 : For each university, extract from the times dataset the most recent and the least recent data, obtaining two separate dataframes

```python
df_rank_times_last_year=df_rank_times_base.loc[df_rank_times_base.groupby(["university_name"])["year"].idxmax()]
df_rank_times_last_year
```

```python
df_rank_times_first_year=df_rank_times_base.loc[df_rank_times_base.groupby(["university_name"])["year"].idxmin()]
df_rank_times_first_year
```

```python
# le università presenti un solo anno in classifica le lasciamo nel file più recente quindi le eliminiamo da questo file meno recente

df_rank_times_first_year=pd.merge(df_rank_times_first_year,df_rank_times_last_year, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
df_rank_times_first_year
```

### 2. For each university, compute the improvement in income between the least recent and the most recent data pointss

```python
df_rank_times_first_last_year_join =df_rank_times_first_year[["university_name","year","income"]].merge(df_rank_times_last_year[["university_name","year","income"]],how='inner',on='university_name')
df_rank_times_first_last_year_join
```

```python
# eliminiamo i casi in cui c'è solo un ranking (non più utile dopo averli leminati già prima) e calcoliamo il miglioramento, 
#e consideriamo solo i casi con income valorizzato  che comunque per definizione dovrebbe essere positivo altrimenti non ha senso
df_rank_times_first_last_year_join_v2=df_rank_times_first_last_year_join[~(df_rank_times_first_last_year_join['year_x']==df_rank_times_first_last_year_join['year_y']) & (df_rank_times_first_last_year_join['income_x'] !='-' ) & (df_rank_times_first_last_year_join['income_y'] !='-' )]

df_rank_times_first_last_year_join_v2['perc_improvement']= ((pd.to_numeric(df_rank_times_first_last_year_join_v2['income_y'],errors='coerce')- pd.to_numeric(df_rank_times_first_last_year_join_v2['income_x'],errors='coerce'))/pd.to_numeric(df_rank_times_first_last_year_join_v2['income_x'],errors='coerce'))*100
df_rank_times_first_last_year_join_v2['improvement']= pd.to_numeric(df_rank_times_first_last_year_join_v2['income_y'],errors='coerce')- pd.to_numeric(df_rank_times_first_last_year_join_v2['income_x'],errors='coerce')
type(df_rank_times_first_last_year_join_v2)
df_rank_times_first_last_year_join_v2
```

### 4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for Aarhus University the value is 122-73=49). Notice that some rankings are expressed as a range

```python
#creo i tre data set con ultimo anno 
df_rank_times_last_year=df_rank_times_base.loc[df_rank_times_base.groupby(["university_name"])["year"].idxmax()]
df_rank_cwur_last_year=df_rank_cwur_base.loc[df_rank_cwur_base.groupby(["university_name"])["year"].idxmax()]
df_rank_shanghai_last_year=df_rank_shanghai_base.loc[df_rank_shanghai_base.groupby(["university_name"])["year"].idxmax()]
```

```python
#seleziono solo le colonne che mi interessano
df_rank_times_last_year_sh=df_rank_times_last_year[['university_name',"year","world_rank_adj"]]
df_rank_cwur_last_year_sh=df_rank_cwur_last_year[['university_name',"year","world_rank_adj"]]
df_rank_shanghai_last_year_sh=df_rank_shanghai_last_year[['university_name',"year","world_rank_adj"]]
```

```python
# outer join di tutte e tre le tabelle --> non facio l'inner nel caso ci sia una tabella mancante e le altre due  valorizzate
# a notare che a volte non si tratta dello stesso anno
# infine ordino per differenza massima

df_rank_outer_join=df_rank_times_last_year_sh.merge(df_rank_cwur_last_year_sh,on='university_name',how="outer").merge(df_rank_shanghai_last_year_sh,on='university_name',how="outer")
df_rank_outer_join["max_rank"]=df_rank_outer_join[["world_rank_adj", "world_rank_adj_y","world_rank_adj_x"]].max(axis=1)
df_rank_outer_join["min_rank"]=df_rank_outer_join[["world_rank_adj", "world_rank_adj_y","world_rank_adj_x"]].min(axis=1)
df_rank_outer_join["max_difference"]=(df_rank_outer_join["max_rank"]-df_rank_outer_join["min_rank"])
df_rank_outer_join=df_rank_outer_join[["university_name","world_rank_adj_x","world_rank_adj_y","world_rank_adj","max_difference"]]
df_rank_outer_join.sort_values(by='max_difference', ascending=False).head(10)
```
